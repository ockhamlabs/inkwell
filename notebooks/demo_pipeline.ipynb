```python
from inkwell.pipeline import Pipeline
from inkwell.io import read_image

from inkwell.utils.visualize import draw_box, visualize_image
```

```python
IMG_PATH = "./data/sample_poster.png"
image = read_image(IMG_PATH)
visualize_image(image)
```

```python
pipeline = Pipeline()
results = pipeline.process(IMG_PATH)


# If you want to parse a PDF document
# PDF_PATH = "./data/sample_poster.pdf"
# pipeline = Pipeline()
# results = pipeline.process(PDF_PATH, pages_to_parse=[0, 1])
```

```python
# Visualize the layouts for the first page
results = results.pages[0]
draw_box(image, layout=results.layout, show_element_type=True)
```

```python
images = results.image_fragments()
tables = results.table_fragments()
text_blocks = results.text_fragments()
```

```python
for fragment in images:
    visualize_image(fragment.content.image)
    print("-"*100)
```

```python
for fragment in tables:
    visualize_image(fragment.content.image)
    # print(f"Data in table:\n{fragment.content.data}")
    print(f"Detection Score: {fragment.content.score}")
    print("-"*100)
```

```python
for text_block in text_blocks:
    visualize_image(text_block.content.image)
    print(f"Text in text block:\n{text_block.content.text}")
    print(f"Detection Score: {text_block.content.score}")
    print("-"*100)
```
