```python
import cv2
from tensordoc.layout_detector import LayoutDetectorFactory, LayoutDetectorType
from tensordoc.utils.visualize import visualize_image
from tensordoc.components.layout import Layout
from layoutparser import draw_box
from IPython.display import display
```

```python
cfg = {"detection_threshold":0.1}


# Load the image
image = cv2.imread("./data/sample.png")
visualize_image(image)
```

```python
def demo_layout_detection(detector_name, image):    
    detector = LayoutDetectorFactory.get_layout_detector(
    detector_name,
    **cfg
    )
    layout = detector.process(image)
    layout = Layout([l for l in layout if l.type in ["Text", "Title"]])
    display(draw_box(image, layout, box_width=3))
```

```python
demo_layout_detection(LayoutDetectorType.FASTER_RCNN, image)
```

```python
demo_layout_detection(LayoutDetectorType.DIT, image)
```

```python
demo_layout_detection(LayoutDetectorType.LAYOUTLMV3, image)
```
