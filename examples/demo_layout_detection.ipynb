```python
import cv2
from matplotlib import pyplot as plt

from tensordoc.layout_detector import LayoutDetectorFactory, LayoutDetectorType
```

```python
# Load the image

image = cv2.imread("./data/sample.png")
image = image[..., ::-1]

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off') 
plt.show()
```

```python
detector = LayoutDetectorFactory.get_layout_detector(LayoutDetectorType.DETECTRON2)
layout = detector.process(image)
```

```python
for l in layout:
    print(l)
```

```python
import layoutparser as lp

lp.draw_box(image, layout, box_width=3)
```
